# Playground

In this Notebook you can easily change parameter values and directly inspect the result of your choices. Upload a portrait of yourself (with the background removed) and choose your favourite settings.

_TIP_: before playing around with the parameters, make sure you know what they do by consulting the Notebook `parameter_demo.ipynb`.

```python
# Import libraries
import matplotlib.pyplot as plt

# Import gol_mosaics (installed with: pip install -e .)
from gol_mosaics import MosaicGenerator, ColorScheme

%load_ext autoreload
%autoreload 2
```

```python
from PIL import Image

# Load input image
input_path = '../input/images/michiel.png'
input_image = Image.open(input_path)

# Display
plt.figure(figsize=(6, 6))
plt.imshow(input_image)
plt.title('Portrait')
plt.axis('off')
plt.show()
```

Choose all your parameter value

```python
level = 4
grid_size = 150
eca_rule = 127
supersample = 12
empty_tiles_cutoff = .75
alpha_cutoff = .8

gol_background = '#1E64C8'
gol_pixel = '#000000'
eca_background = '#FD7762'
eca_pixel = '#1E64C8'

color_scheme = ColorScheme.custom(gol_background = gol_background,
                                  gol_pixel = gol_pixel,
                                  eca_background = eca_background,
                                  eca_pixel = eca_pixel)
```

... and generate the image

```python
generator = MosaicGenerator(level=level,
                            grid_size=grid_size,
                            eca_rule=eca_rule,
                            color_scheme=color_scheme)

mosaic = generator.generate_from_image(input_path,
                            supersample=supersample,
                            empty_tiles_cutoff=empty_tiles_cutoff,
                            alpha_cutoff=alpha_cutoff)

fig, ax = plt.subplots(1,1, figsize=(10, 10))

ax.imshow(mosaic)
ax.axis('off')
```

Like it? If not, try some other setting :) Otherwise, you can save it below

```python
SAVEFIG = True

if SAVEFIG:
    output_path = '../output/playground.png'
    mosaic.save(output_path)
else:
    print("Mosaic not saved. Set SAVEFIG = True to save the Mosaic to the output directory.")
```
